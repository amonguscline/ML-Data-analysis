# Regression


Regressio should be used when you want to predict a value of the target column over a range of continuous values. Regression will give you a correlation line basically. Target columns might be height, distance, revenue or population

#Classification

In classification, we predict whether or not a target value fits into a category. Usually classification sorts into either 0 or 1. Target columns might be color labels, OCR, etc. and we can also split a continous value into a true or false value, such as is the value greater than a certain number

#Bike Rentals with Decision Tree Regressor

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
```

```python
url="https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/refs/heads/master/Chapter03/bike_rentals_cleaned.csv"
df=pd.read_csv(url)
y=df["cnt"] #my dataset doesn't have casual
X=df.loc[:,"season":"windspeed"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
model = DecisionTreeRegressor(max_depth=5)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

#Different models

```python
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
def regressor(model):
  X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
  model.fit(X_train,y_train)
  return model.score(X_test,y_test)
```

```python
regressor(LinearRegression())
```

```python
regressor(Ridge(alpha=.5))
```

```python
regressor(KNeighborsRegressor(5))
```

```python
regressor(AdaBoostRegressor())
```

```python
regressor(MLPRegressor())
```

#KNN

KNN groups an X value based on the K nearest neigbors' X values and returns the mean of the K nearest neighbors' y values. Same works with classification. We usually use an odd value for K in regression because we'll usually have a mean. KNN is good because it's easy to interpret and because no parameters are required, but it's bad because models might be less effective with more dimensions and because it doesn't usually outperform parametric models.

```python
for i in range(1,10):
  print(regressor(KNeighborsRegressor(n_neighbors=i)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
url="https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/refs/heads/master/Chapter06/heart_disease.csv"
df= pd.read_csv(url)
y=df.iloc[:,-1]
X=df.iloc[:,:-1]
scaler = preprocessing.MinMaxScaler()
X_normalized = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_train,X_test,y_train,y_test=train_test_split(X_normalized,y)
```

```python
for i in range(1,10):
  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(X_train,y_train)
  print(f'{i }:{model.score(X_test,y_test)}')
```
