Deep learning is more mathematically intricate than machine learning, but it performs better with unstructured data and very large datasets.
In a two-layer neural network, it works the same as linear regression by multiplying the inputs by the weights and adding a bias. Then you use gradient descent to minimize error.
With logistic regression, it's the same except you use the sigmoid function to predict y based on X

It's more common in neural nets to have a hidden layer between the inputs and the outputs. The hidden layers increase complexity in models because of the increased weights and biases that can affect the cost function.

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping
from IPython.display import clear_output
```

```python
df_train= pd.read_csv("/content/sample_data/california_housing_train.csv")
df_test =pd.read_csv("/content/sample_data/california_housing_test.csv")
```

```python
y=df_train.iloc[:,-1]
X=df_train.iloc[:,:-1]
```

```python
model = Sequential()

#add a hidden layer with all columns and relu activation to force positives
model.add(Dense(50,input_shape=(X.shape[1],), activation='relu'))
model.add(Dropout(.2)) #drops nodes to prevent overfitting
model.add(Dense(25, activation='relu'))
model.add(Dense(10, activation='relu'))

model.add(Dense(1, activation='relu'))

model.summary()
```

```python
model.compile(optimizer='adam',loss='mse')
#patience is how many epochs we go before it gets impatient and stops
esm=EarlyStopping(patience=50)
model.fit(X,y,epochs=1000,validation_split=.1,callbacks=[esm])
mse=model.evaluate(df_test.iloc[:,:-1],df_test.iloc[:,-1])
clear_output()
mse**.5
```
