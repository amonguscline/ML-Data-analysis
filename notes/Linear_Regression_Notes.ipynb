```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df
```

Data needs to have no null values and all cells need to be numeric values for ML

```python
y=df['median_house_value'] # output
X=df.iloc[:,:-1] #inputs
```

```python
#linear regression plot
sns.set()
sns.regplot(data=df, x=df['total_rooms'], y=y)
```

In 3d you get a hyperplane instead of a line

# machine learning

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y) #r squared
```

```python
model2 = LinearRegression(fit_intercept=False)
model2.fit(X,y)
model2.score(X,y)
```
