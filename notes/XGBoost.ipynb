#Decision Trees versus XGBoost


Random forests don't improve with each iteration, so it can be a bad method if the trees aren't very good. By contrast, gradient boosting aims to improve the trees each iteration. Gradient boosting works as an ensemble of decision trees, but these trees evolve each iteration by training the new trees on the error of the previous tree. This way, each subsequent tree will improve on gains made by previous trees, turning a weak learner into a stronger learner. Overall, gradient boosting is more effective than bagging that is used in random forests. Boosting is a sequential process, whereas bagging is a parallel process.

Gradient boosting uses residuals (the difference between the predictions and the test set aka the errors) to train the model in order to get closer to a regression line. Gradient boosting can be improved with XGBoost, which is a very efficient model for gradient boosting.

```python
import pandas as pd
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from ucimlrepo import fetch_ucirepo
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
heart_disease = fetch_ucirepo(id=45)
X = heart_disease.data.features
y = heart_disease.data.targets
y
```

```python
def classifier(model):
  scores=cross_val_score(model,X,y,cv=5,scoring="accuracy")
  return scores.mean()
```

```python
model = XGBClassifier()
classifier(model)
```
