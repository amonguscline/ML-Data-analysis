```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

#logistic regression

```python
df = pd.read_csv('https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/refs/heads/master/Chapter03/census_cleaned.csv')
y=df.iloc[:,-1]
X=df.iloc[:,:-1]
model=LogisticRegression(max_iter=1000, C=1.0, fit_intercept=False) #max iter limits time it will take to run the code
#warm_start=True can make it so that you can run more iterations and builds off the last model.fit()
model.fit(X,y)
model.score(X,y)
```


logistic regression is a classifier, meaning that it predicts if something is true or false by classifying data into two groups
it starts by doing the same back propogation process, but then it just plots the points with x and y by plugging and then classifies each data point as class 1 or class 0 (boolean)
values near the middle of the sigmoid function that classifies the data points are more uncertain because the ML model could classify them either way
the y value is interpereted as the probability of that point being classified.
y=1/1-e^+t is the sigmoid, where t is X
logistic regression should be used to classify data based on a y boolean value (ranging from 0 to 1)

```python
model.predict_proba(X.iloc[-5:,:])
```

#Math behind logistic regression


We use a sigmoid function, 1/1+e^-X, which has asymptotes of y=1 and y=0
this sigmoid acts as a "classifier" because it is evenly split over X<0 and X>0, so it will either sort our X values into a "true" bin or a "false" bin.
the y value that is output from plugging in X (sum of the columns and their weights) is our probability that X should be in that bin
