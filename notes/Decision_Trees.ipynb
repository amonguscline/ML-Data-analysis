#Decision Trees
The data is split into two groups based on which split gives the best score.
This process is continued until it's as accurate as you want it to be based on max_depth or if it converges.
The data is split like a binary tree, so it's either split into class 1 or 0, true or false.
The gini value given by our decision tree is similar to the probability, but it's kind of reversed (probability is 1-gini, so .01 is a pretty high probability).
Decision trees are a form of non-linear machine learning.

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #Classification; we're predicting a boolean value
```

```python
url = "https://media.githubusercontent.com/media/PacktPublishing/Hands-On-Gradient-Boosting-with-XGBoost-and-Scikit-learn/master/Chapter02/heart_disease.csv"
df=pd.read_csv(url)
y=df.iloc[:,-1]
X=df.iloc[:,:-1]
```

```python
model = DecisionTreeClassifier()
model.fit(X,y)
model.score(X,y) #don't score based on the X and y that we used to make the model, the data is 1:1
```

#Overfitting
Linear regression lines can fail to accurately represent our data, resulting in underfitting.
We can use higher degree polynomials that will accurately pick up on all of our data, which sounds good but then you realize that if we add more data, the curve give us a less accurate depiction of the data. This is called overfitting.
Overfitting can also bias outliers, which we don't want.
We want to find a curve that doesn't underfit or overfit, which can generalize to new data that it wasn't trained on.
To solve overfitting, we want to split the data into a training and a test set. We can also use models with built in regularization, which will automatically prevent overfitting. You can also cross validate, which trains the model on multiple sections of data, or you can tweak some of the hyperparameters (like max_depth)

```python
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into training and testing data to reduce overfitting
X_train, X_test, y_train, y_test = train_test_split(X,y)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn import tree
model.tree_.max_depth
```

```python
#changing hyperparameters
model = DecisionTreeClassifier(max_depth=3)
#max_depth changes how many splits our trees, which can generalize the data better at lower numbers
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model = DecisionTreeClassifier(max_depth=3,min_samples_leaf=4)
#min_samples_leaf splits until we get to a certain number of leaves, which can give more generality
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

We can also change the paremeters of train_test_split, by using test_size=% of data (floating point number) to reduce overfitting.
