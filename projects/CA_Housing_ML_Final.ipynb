```python
houses=pd.read_csv('/content/sample_data/california_housing_train.csv')
filteredhouses = houses[(houses['median_house_value']<500001)]
houses_X=houses.iloc[:,:-1]
houses_y=houses.iloc[:,-1]
filtered_X=filteredhouses.iloc[:,:-1]
filtered_y=filteredhouses.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
```

```python
def regressor(model,X,y):
  return (-cross_val_score(model,X,y,scoring="neg_mean_squared_error").mean())**(1/2)
```

```python
def regressor2(params,model,X,y,random=False):
  if random:
    regressor=RandomizedSearchCV(model,params,n_iter=20,scoring="neg_mean_squared_error")
  else:
    regressor = GridSearchCV(model,params,scoring="neg_mean_squared_error")
  regressor.fit(X,y)
  best_params=regressor.best_params_
  best_score=regressor.best_score_
  print(best_params)
  return((-best_score)**(1/2))
```

```python
def pretty_print(mean,model_type):
  print(f'{model_type}: {mean}')
```

```python
pretty_print(regressor(LinearRegression(),houses_X,houses_y),"Linear Regression")
pretty_print(regressor(KNeighborsRegressor(),houses_X,houses_y),"KNN Regressor")
pretty_print(regressor(RandomForestRegressor(),houses_X,houses_y),"Random Forest Regressor")
pretty_print(regressor(XGBRegressor(),houses_X,houses_y),"XGBoost")
```

```python
pretty_print(regressor(LinearRegression(),filtered_X,filtered_y),"Linear Regression")
pretty_print(regressor(KNeighborsRegressor(),filtered_X,filtered_y),"KNN Regressor")
pretty_print(regressor(RandomForestRegressor(),filtered_X,filtered_y),"Random Forest Regressor")
pretty_print(regressor(XGBRegressor(),filtered_X,filtered_y),"XGBoost")
```

```python
LR_params={
    'fit_intercept':[True,False]
}
pretty_print(regressor2(LR_params,LinearRegression(),filtered_X,filtered_y,random=False),"Linear Regression")
```

```python
KNN_params={
    'n_neighbors':[5,6,8,10,12,14,20],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size':[10,20,30,40,50,70]
}
pretty_print(regressor2(KNN_params,KNeighborsRegressor(),filtered_X,filtered_y,random=False),"KNN Regressor")
```

```python
RF_params={
   'n_estimators':[40,50,60,70],
   'max_depth':[8,10,12],
   'min_samples_split':[4,5,6],
   'min_samples_leaf':[1,2],
   'max_features':[.4,.6],
   'min_impurity_decrease':[0,.5,1],
   'ccp_alpha':[.2,.3]
}
pretty_print(regressor2(RF_params,RandomForestRegressor(),filtered_X,filtered_y,random=True),"Random Forest Regressor")
```

```python
XGB_params={
    'max_depth':[2,3],
    'subsample':[.8,.9],
    'colsample_bytree':[.9],
    'colsample_bynode':[.9],
    'colsample_bylevel':[.9],
    'learning_rate':[.2,.3],
    'gamma':[.5,.55],
    "n_estimators":[80]
}
pretty_print(regressor2(XGB_params,XGBRegressor(),filtered_X,filtered_y,random=False),"XGBoost")
```
