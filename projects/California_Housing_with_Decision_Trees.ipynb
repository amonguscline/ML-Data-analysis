```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
test=pd.read_csv("/content/sample_data/california_housing_test.csv")
test
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
model = DecisionTreeRegressor(max_depth=7,min_samples_leaf=8)
model.fit(X,y)
#using the built in test csv from google colab
test_X=test.iloc[:,:-1]
test_y=test.iloc[:,-1]
model.score(test_X,test_y)
```
