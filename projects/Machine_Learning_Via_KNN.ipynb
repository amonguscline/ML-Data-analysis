```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
y=df.iloc[:,-1]
X=df.iloc[:,:-1]
scaler=MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
for i in range(1,20):
  model= KNeighborsRegressor(n_neighbors=i)
  model.fit(X_train,y_train)
  print(model.score(X_test,y_test))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1)
```

```python
for i in range(1,20):
  model=KNeighborsRegressor(n_neighbors=i)
  model.fit(X_train,y_train)
  print(model.score(X_test,y_test))
```

I think that KNN is a decent model for my dataset because a lot of the neighbors of an individual X will be pretty accurate in terms of predicting that X, just like how houses in one neighborhood might be good at predicting the houses of their neighbors.
