```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("/content/sample_data/california_housing_test.csv")

bayarea=pd.DataFrame()
bayarea["longitude"]=df.iloc[:,0].map(lambda s:1 if s>-122.644 and s<-121.587 else 0)
bayarea["latitude"]=df.iloc[:,1].map(lambda s:1 if s>37.189 and s<38.203 else 0)
def combineRows(a,b):
    return a and b
df["in_bay_area"]=bayarea.iloc[:,0:2].apply(lambda row:combineRows(row["longitude"],row["latitude"]),axis=1)
X=df.iloc[:,2:-1]
y=df["in_bay_area"]
model = LogisticRegression(max_iter=1000)

model.fit(X,y)
model.score(X,y)
sample=df.sample(10)
model.predict_proba(sample.iloc[:,2:-1])
```
