```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def fit(data_frame):
  model = LinearRegression()
  y=data_frame["median_house_value"] #defining x and y for ML
  X=data_frame.iloc[:,0:-1]
  model.fit(X,y)
  return model
```

```python
def score (model, test):
  testy=test["median_house_value"]
  testX=test.iloc[:,0:-1]
  return model.score(testX,testy)
```

```python
df = pd.read_csv('./sample_data/california_housing_train.csv')
test = pd.read_csv('./sample_data/california_housing_test.csv')
test["median_house_value"].describe()
```

```python
model=fit(df)
score(model, test)
```

#First attempt at improving data
##truncated values of median house value <500001 >14999

```python
filtereddf = df[(df['median_house_value']<500001) & (df["median_house_value"]>0)]
filteredmodel=fit(filtereddf)
score(filteredmodel, test)
```

```python
filteredtest=test[(test['median_house_value']<500001) & (test["median_house_value"]>0)]
score(filteredmodel,filteredtest)
```
