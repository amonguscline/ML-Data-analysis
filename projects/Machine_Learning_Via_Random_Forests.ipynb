```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
from sklearn.model_selection import cross_val_score
```

```python
url="/content/sample_data/california_housing_train.csv"
df=pd.read_csv(url)
df=df[df["median_house_value"]<500000] #remove outliers
df.describe()
y=df.iloc[:,-1]
X=df.iloc[:,:-1]
```

```python
model=RandomForestRegressor(max_depth=8,n_estimators=100,min_samples_leaf=4,max_samples=.8)
(-cross_val_score(model,X,y,scoring='neg_mean_squared_error').mean())**(1/2)
```
